```python
from transformers import pipeline

qa_model = pipeline("question-answering")
```

```python
context = """
Diabetes is a chronic disease that occurs when the pancreas does not produce enough insulin
or when the body cannot effectively use the insulin it produces. Insulin is a hormone that
regulates blood sugar. Over time, uncontrolled diabetes can damage the heart, blood vessels,
eyes, kidneys, and nerves. Maintaining a healthy diet, regular physical activity, and a normal
body weight are key to preventing diabetes.
"""
```

```python
question = "What does insulin regulate?"
result = qa_model(question=question, context=context)
print(result['answer'])
```

```python
questions = [
    "What disease is caused by lack of insulin?",
    "What organ produces insulin?",
    "How can diabetes be prevented?"
]

for q in questions:
    answer = qa_model(question=q, context=context)
    print(f"Q: {q}\nA: {answer['answer']}\n")
```
